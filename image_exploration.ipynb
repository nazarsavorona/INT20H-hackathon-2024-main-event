```python
import os
import random
import cv2
import matplotlib.pyplot as plt
```

```python
images_folder = "faces"
image_paths = [os.path.join(images_folder, img) for img in os.listdir(images_folder) if img.endswith(".jpg")]

random.seed(42)
random_images = random.sample(image_paths, 10) 
```

```python
plt.figure(figsize=(15, 6))
for i, image_path in enumerate(random_images):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    red_channel = image[:, :, 0].flatten()
    green_channel = image[:, :, 1].flatten()
    blue_channel = image[:, :, 2].flatten()
    
    plt.subplot(10 // 5, 5, i + 1)
    plt.hist(red_channel, bins=256, color='red', alpha=0.5, label='Red')
    plt.hist(green_channel, bins=256, color='green', alpha=0.5, label='Green')
    plt.hist(blue_channel, bins=256, color='blue', alpha=0.5, label='Blue')
    
    # set xlabel and ylabel for subplot
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    
    plt.title(f'Image {i + 1}')
    plt.axis('off')
    
plt.tight_layout()
plt.show()

# plot corresponding images
plt.figure(figsize=(15, 6))
for i, image_path in enumerate(random_images):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    plt.subplot(10 // 5, 5, i + 1)
    plt.imshow(image)
    plt.title(f'Image {i + 1}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 6))
for i, image_path in enumerate(random_images):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image_hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
    
    hue_channel = image_hsv[:, :, 0].flatten()
    saturation_channel = image_hsv[:, :, 1].flatten()
    value_channel = image_hsv[:, :, 2].flatten()
    
    plt.subplot(10 // 5, 5, i + 1)
    plt.hist(hue_channel, bins=256, color='red', alpha=0.5, label='Hue')
    plt.hist(saturation_channel, bins=256, color='green', alpha=0.5, label='Saturation')
    plt.hist(value_channel, bins=256, color='blue', alpha=0.5, label='Value')
    
    # set xlabel and ylabel for subplot
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    
    plt.title(f'Image {i + 1}')
    plt.axis('off')
    
plt.tight_layout()
plt.show()

# plot corresponding images
plt.figure(figsize=(15, 6))
for i, image_path in enumerate(random_images):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    plt.subplot(10 // 5, 5, i + 1)
    plt.imshow(image)
    plt.title(f'Image {i + 1}')
    plt.axis('off')
    
plt.tight_layout()
plt.show()
```
